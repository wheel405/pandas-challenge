```python
#print data table with summary of school district
total_schools = school_data['school_name'].count()
total_students =student_data['Student ID'].count()
total_budget = school_data['budget'].sum()
average_math_score = round(student_data['math_score'].mean())
average_reading_score = round(student_data['reading_score'].mean())
overall_average_score = round((average_math_score + average_reading_score)/2)
passing_math = round(school_data_complete.query('math_score >70')["School ID"].count()/ total_students*100)
passing_reading = round(school_data_complete.query('reading_score >70')["School ID"].count()/ total_students*100)
school_dicts = [{"Total Schools": (total_schools),"Total Students": (total_students), 
                 "Total Budget": (total_budget), "Average Math Score": (average_math_score),
                 "Average Reading Score": (average_reading_score), "Overall Average Score": (overall_average_score),
                "% of Students Passing Math": (passing_math), "% of Students Passing Reading ": (passing_reading)}]
district_summary_df = pd.DataFrame(school_dicts)
district_summary_df


```

```python
#group by school name and type. use first function to delete the unnecessary brackets around [district] or [charter]
school_type = school_data_complete.groupby(['school_name'])['type'].first()


school_type.head()
```

```python
total_students_per_school= school_data_complete['school_name'].value_counts()
total_students_per_school
```

```python
total_school_budget=school_data_complete.groupby('school_name')['budget'].first()
total_school_budget
```

```python
per_student_budget = total_school_budget/total_students_per_school
per_student_budget
```

```python
avg_math_score = (school_data_complete.groupby(["school_name"]).math_score.mean())
avg_math_score
```

```python
avg_read_score = (school_data_complete.groupby(["school_name"]).reading_score.mean())
avg_read_score
```

```python
#% passing math
pct70_mathpass = school_data_complete.loc[school_data_complete['math_score'] >= 70,:].groupby(['school_name']).size()
passing_math_by_school = (pct70_mathpass/total_students_per_school)*100
passing_math_by_school
```

```python
#% passing reading
pct70_readpass = school_data_complete.loc[school_data_complete['reading_score'] >= 70,:].groupby(['school_name']).size()
passing_reading_by_school = ((pct70_readpass/total_students_per_school)*100)
passing_reading_by_school
```

```python
#Overall Passing Rate (Average of the above two)
overall_pass_rate = (passing_math_by_school + passing_reading_by_school)/ 2
overall_pass_rate
```

```python
#create a dictionary and a dataframe to view school summary data
school_summary_dicts ={"School Type": school_type,"Total Students": total_students_per_school,
                               "Total School Budget": total_school_budget,"Per Student Budget": per_student_budget,
                                "Average Math Score": avg_math_score,"Average Reading Score": avg_read_score,
                                "% Passing Math": passing_math_by_school,"% Passing Reading": passing_reading_by_school,
                                "% Overall Passing Rate": overall_pass_rate}
school_summary_df = pd.DataFrame(school_summary_dicts)
school_summary_df
```

```python
#format and clean the data from the table above
School_Summary = pd.DataFrame(school_summary_df)
formatted_school_summary = School_Summary.copy()
formatted_school_summary['Total School Budget'] = formatted_school_summary['Total School Budget'].map('${:,.2f}'.format)
formatted_school_summary['Per Student Budget'] = formatted_school_summary['Per Student Budget'].map('${:,.2f}'.format)
formatted_school_summary
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe

```python

#Calculate the average math score by grade 9 per school
avg_math_school_9grade = school_data_complete.loc[school_data_complete.grade == '9th'].groupby('school_name')['math_score'].mean()

#Calculate the average math score by grade 10 per school
avg_math_school_10grade = school_data_complete.loc[school_data_complete.grade == '10th'].groupby('school_name')['math_score'].mean()

#Calculate the average math score by grade 11 per school
avg_math_school_11grade = school_data_complete.loc[school_data_complete.grade == '11th'].groupby('school_name')['math_score'].mean()

#Calculate the average math score by grade 12 per school
avg_math_school_12grade = school_data_complete.loc[school_data_complete.grade == '12th'].groupby('school_name')['math_score'].mean()

#Create a dictionary to hold values:
avg_math_dicts = {'9th Grade': avg_math_school_9grade,
         '10th Grade': avg_math_school_10grade,
           '11th Grade': avg_math_school_11grade,
           '12th Grade': avg_math_school_12grade}

#create dataframe to display values in the above dictionary
Avg_Math_School_Grade = pd.DataFrame(avg_math_dicts)
Avg_Math_School_Grade
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:

  * Average Reading Score

```python
# Bins made based on school_summary table before
spending_bins = [0, 599, 615, 649, 700]
group_names = ["<$599", "$599-615", "$615-649", "$649-700"]
```

```python
#Create new dataframe with columns to match the bullet pointed list of stuff requested in the above instructions
scores_by_school_spending = School_Summary.loc[:,['Average Math Score',
                                           'Average Reading Score',
                                          '% Passing Math',
                                          '% Passing Reading',
                                          '% Overall Passing Rate',]]

#Add a new column called Spending Ranges= then bin based off budget per student
scores_by_school_spending['Spending Ranges (Per Student)'] = pd.cut(School_Summary['Per Student Budget'], spending_bins,labels=group_names)

#display dataframe using ('Spending Ranges (Per Student)') as column header
scores_by_spending = scores_by_school_spending.groupby('Spending Ranges (Per Student)').mean()

scores_by_spending
```
